```python
import requests
import pandas as pd
import folium # map rendering library
from geopy.geocoders import Nominatim
import math  
    
request = requests.get('https://api.covid19.finspect.me/brcovid19/map')
df=pd.DataFrame(request.json())
address = 'Brazil'
 
geolocator = Nominatim(user_agent="CovidBrazil")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
 
# create map of New York using latitude and longitude values
map_Covid = folium.Map(location=[latitude, longitude], zoom_start=4)

df["lat"] = pd.to_numeric(df["lat"])
df["lng"] = pd.to_numeric(df["lng"])
 
# add markers to map
for lat, lng, borough, neighborhood in zip(df['lat'], df['lng'], df['name'], df['qtd_confirmation']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=neighborhood/500,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Covid)  
    
map_Covid
```

```python
map_Covid.save('index.html')
```
